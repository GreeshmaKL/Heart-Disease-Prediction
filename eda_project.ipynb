```python
le=LabelEncoder()
df['Heart Disease']=le.fit_transform(df['Heart Disease'])
scaler=StandardScaler()
x=scaler.fit_transform(df.drop(['Heart Disease'],axis=1))
y=df['Heart Disease']
```

```python
x=df.drop(['Heart Disease'],axis=1)
y=df['Heart Disease']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.2)
```

```python
# Decision Tree algorithm
from sklearn.tree import DecisionTreeClassifier
# Creating a decision tree classifier
dt = DecisionTreeClassifier(max_depth=3)

# Training the classifier
dt.fit(x_train, y_train)

# Predicting the target values
y_pred1 = dt.predict(x_test)
cm1=confusion_matrix(y_test,y_pred1)
print(cm1)
# Evaluating the accuracy of the model
accuracy1 = accuracy_score(y_test, y_pred1)
print("Accuracy using Decision Tree algorithm is ",accuracy1)
score1=f1_score(y_test,y_pred1)
print("F1 score of using Decision Tree algorithm is ",score1)
```

```python
# Random forest
from sklearn.ensemble import RandomForestClassifier

# Training the algorithm
rfc=RandomForestClassifier(n_estimators=100,random_state=42)
rfc.fit(x_train,y_train)

# Make prediction on testing data
y_pred2=rfc.predict(x_test)
cm2=confusion_matrix(y_test,y_pred2)
print(cm2)
accuracy2=accuracy_score(y_test,y_pred2)
print("Accuracy using Random Forest Classifier is ",accuracy2)
score2=f1_score(y_test,y_pred2)
print("F1 score of using Random Forest Classifier is ",score2)
```

```python
# Logistic Regression
from sklearn.linear_model import LogisticRegression
# Training
lrc=LogisticRegression(max_iter=10000)
lrc.fit(x_train,y_train)
# Testing
y_pred3=lrc.predict(x_test)
cm3=confusion_matrix(y_test,y_pred3)
print(cm3)
accuracy3=accuracy_score(y_test,y_pred3)
print("Accuracy using Logistic Regression is ",accuracy3)
score3=f1_score(y_test,y_pred3)
print("F1 score of using Logistic Regression is ",score3)
```

```python
# Support Vector Machine linear kernel
from sklearn.svm import SVC
# Training
svm=SVC(kernel='linear')
svm.fit(x_train,y_train)
# Testing
y_pred41=svm.predict(x_test)
cm41=confusion_matrix(y_test,y_pred41)
print(cm41)
accuracy41=accuracy_score(y_test,y_pred41)
print("Accuracy using Support Vector Machine and linear kernel is ",accuracy41)
score41=f1_score(y_test,y_pred41)
print("F1 score of using Support Vector Machine and linear kernel is ",score41)
```

```python
# Support Vector Machine rbf kernel
from sklearn.svm import SVC
# Training
svm=SVC(kernel='rbf')
svm.fit(x_train,y_train)
# Testing
y_pred42=svm.predict(x_test)
cm42=confusion_matrix(y_test,y_pred42)
print(cm42)
accuracy42=accuracy_score(y_test,y_pred42)
print("Accuracy using Support Vector Machine and rbf kernel is ",accuracy42)
score42=f1_score(y_test,y_pred42)
print("F1 score of using Support Vector Machine and rbf kernel is ",score42)
```

```python
# Support Vector Machine polynomial kernel
from sklearn.svm import SVC
# Training
svm=SVC(kernel='poly',degree=3)
svm.fit(x_train,y_train)
# Testing
y_pred43=svm.predict(x_test)
cm43=confusion_matrix(y_test,y_pred43)
print(cm43)
accuracy43=accuracy_score(y_test,y_pred43)
print("Accuracy using Support Vector Machine and polyomial kernel of degree 4 is ",accuracy43)
score43=f1_score(y_test,y_pred43)
print("F1 score of using Support Vector Machine and polynomial kernel of degree 4 is ",score43)
```

```python
# KNN
from sklearn.neighbors import KNeighborsClassifier
# Training
knn=KNeighborsClassifier(n_neighbors=11)
knn.fit(x_train,y_train)
# Testing
y_pred5=knn.predict(x_test)
cm5=confusion_matrix(y_test,y_pred5)
print(cm5)
accuracy5=accuracy_score(y_test,y_pred5)
print("Accuracy using K-Nearest Neighbors is ",accuracy5)
score5=f1_score(y_test,y_pred5)
print("F1 score of using K-Nearest Neighbors is ",score5)
```

```python
# XGBoost algorithm
import xgboost as xgb
xgb_model = xgb.XGBClassifier(n_estimators=100, learning_rate=0.1, max_depth=3)
# training
xgb_model.fit(x_train, y_train)
# testing
y_pred6 = xgb_model.predict(x_test)
cm6=confusion_matrix(y_test,y_pred6)
print(cm6)
accuracy6=accuracy_score(y_test,y_pred6)
print("Accuracy using XGBoost algorithm is ",accuracy6)
score6=f1_score(y_test,y_pred6)
print("F1 score of using XGBoost algorithm is ",score6)
```

```python
# Gaussian Naive Bayes Classifier
from sklearn.naive_bayes import GaussianNB
# Training
gnb=GaussianNB()
gnb.fit(x_train,y_train)
# Testing
y_pred71=gnb.predict(x_test)
cm71=confusion_matrix(y_test,y_pred71)
print(cm71)
accuracy71=accuracy_score(y_test,y_pred71)
print("Accuracy using Gaussian Naive Bayes Classifier is ",accuracy71)
score71=f1_score(y_test,y_pred71)
print("F1 score of using Gaussian Naive Bayes Classifier is ",score71)
```

```python
# Multinomial Naive Bayes Classifier
from sklearn.naive_bayes import MultinomialNB
# Training
mnb=MultinomialNB()
mnb.fit(x_train,y_train)
# Testing
y_pred72=mnb.predict(x_test)
cm72=confusion_matrix(y_test,y_pred72)
print(cm72)
accuracy72=accuracy_score(y_test,y_pred72)
print("Accuracy using Multinomial Naive Bayes Classifier is ",accuracy72)
score72=f1_score(y_test,y_pred72)
print("F1 score of using Multinomial Naive Bayes Classifier is ",score72)
```

```python
# ANN - Artificial Neural Network
from keras.models import Sequential
from keras.layers import Dense
# Define the ANN model
classifier = Sequential()
classifier.add(Dense(activation = "relu", input_dim = 13,units = 8, kernel_initializer = "uniform"))
classifier.add(Dense(activation = "relu", units = 14,kernel_initializer = "uniform"))
classifier.add(Dense(activation = "sigmoid", units = 1,kernel_initializer = "uniform"))
classifier.compile(optimizer = 'adam' , loss = 'binary_crossentropy',metrics = ['accuracy'] )
```

```python
classifier.fit(x_train,y_train,batch_size=8,epochs=100)
```

```python
y_pred8=classifier.predict(x_test)
y_pred8=(y_pred8 > 0.5)
```

```python
cm8=confusion_matrix(y_test,y_pred8)
print(cm8)
accuracy8=accuracy_score(y_test,y_pred8)
print("Accuracy using ANN is ",accuracy8)
score8=f1_score(y_test,y_pred8)
print("F1 score of using ANN is ",score8)
```

```python
classifiers_used=['DT','RF','LR','SVM(L)','SVM(R)','SVM(P)','KNN','XGBoost','GNB','MNB','ANN']
accuracies=[accuracy1,accuracy2,accuracy3,accuracy41,accuracy42,accuracy43,accuracy5,accuracy6,accuracy71,accuracy72,accuracy8]
F1_scores=[score1,score2,score3,score41,score42,score43,score5,score6,score71,score72,score8]
```

```python
# Plot the accuracies
plt.bar(classifiers_used, accuracies)
plt.ylim(0, 1)
plt.title("Accuracy of Different Algorithms for Heart Disease Detection")
plt.xlabel("Algorithm")
plt.ylabel("Accuracy")
plt.show()
```
